## Implementation of IPTW
1ï¸âƒ£ Estimate Propensity Scores

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import numpy as np

# Select pre-treatment variables (confounders)
X = df[["index_age", "total_pre_index_cannisters_365", "pre_asthma_charge", "female", "pneumonia", "gerd"]]
y = df["drug_s"]  # Treatment indicator (0 = Drug_D, 1 = Drug_S)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit logistic regression model to estimate propensity scores
model = LogisticRegression()
model.fit(X_scaled, y)

# Get propensity scores (probability of receiving Drug_S)
df["propensity_score"] = model.predict_proba(X_scaled)[:, 1]
```

```python
# Compute weights
df["iptw_weight"] = np.where(df["drug_s"] == 1, 1 / df["propensity_score"], 1 / (1 - df["propensity_score"]))

# Check weight distribution
df["iptw_weight"].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot distribution of weights
plt.figure(figsize=(8, 6))
sns.histplot(df["iptw_weight"], bins=30, kde=True)
plt.title("Distribution of IPTW Weights")
plt.show()

# Compare pre-treatment characteristics before & after weighting
for var in ["index_age", "total_pre_index_cannisters_365", "pre_asthma_charge"]:
    print(f"Mean {var} before weighting (Drug_S=1):", df[df["drug_s"] == 1][var].mean())
    print(f"Mean {var} after weighting (Drug_S=1):", np.average(df[var], weights=df["iptw_weight"], axis=0))
```

```python
# Compute weighted means for each drug group
mean_exacerbations_weighted = np.average(df["post_index_exacerbations365"], weights=df["iptw_weight"], axis=0)

print(f"Weighted Mean Exacerbations After IPTW: {mean_exacerbations_weighted:.2f}")
```

```python
import statsmodels.api as sm

# Run weighted regression
X = df[["drug_s"]]
X = sm.add_constant(X)
y = df["post_index_exacerbations365"]

# Apply weights
model = sm.WLS(y, X, weights=df["iptw_weight"]).fit()
print(model.summary())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define pre-treatment variables to check for imbalance
pre_treatment_vars = ["index_age", "total_pre_index_cannisters_365", "pre_asthma_charge"]

plt.figure(figsize=(12, 6))
for i, var in enumerate(pre_treatment_vars, 1):
    plt.subplot(1, 3, i)
    sns.kdeplot(df[df["drug_s"] == 0][var], label="Drug_D (0)", shade=True, color="red")
    sns.kdeplot(df[df["drug_s"] == 1][var], label="Drug_S (1)", shade=True, color="blue")
    plt.title(f"Pre-Treatment Distribution of {var}")
    plt.legend()
plt.tight_layout()
plt.show()
```
